```python
import matplotlib.pyplot as plt # Visualization
```

```python
import seaborn as sns # Distribution and Advanced Visualization
```

```python
%matplotlib inline
plt.rcParams['figure.figsize'] = 10,4
```

```python
import warnings
warnings.filterwarnings('ignore')
```

```python
vis1 = sns.distplot(df["InternetUsers"]) # univariate analysis
plt.show(vis1) # it is advanced graph
```

```python
vis1 = sns.displot(df["InternetUsers"]) # t is removed. so distribution line is removed.
plt.show()
```

```python
vis2 = sns.boxplot(data = df, x= 'IncomeGroup', y = 'BirthRate')
plt.show()
```

```python
# outlier = anomaly detection
# Its very very important in machine learning
# outliers are nothing but they are far from other observations.
# they are not the neihgbours, they stay far from others
# the diamomd or circled shaped in boxplot graph is an outlier
# Blue box = interquartile range [IQR]
# midline is called average lines
# lower line  = minimum
# Upper line = maximum 
```

```python
vis3 = sns.lmplot(data=df, x = "InternetUsers", y = "BirthRate")
plt.show() # the blue dots are each data present in the excel
```

```python
vis3 = sns.lmplot(data=df, x = "InternetUsers", y = "BirthRate", fit_reg = False)
plt.show() 
```

```python
vis3 = sns.lmplot(data=df, x = "InternetUsers", y = "BirthRate", fit_reg = False, hue = 'IncomeGroup')
plt.show() # "hue" is same as "legend"
```

```python
vis3 = sns.lmplot(data=df, x = "InternetUsers", y = "BirthRate", fit_reg = True, hue = 'IncomeGroup')
plt.show()
```
