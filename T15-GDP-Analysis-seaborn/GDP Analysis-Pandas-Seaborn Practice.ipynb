```python
vis1 = sns.distplot(gdp["InternetUsers"]) # distplot = distribution plot
plt.show()
```

```python
vis2 = sns.boxplot(data = gdp, x="IncomeGroup", y = "BirthRate")
plt.show()
# boxplot() is a function in Seaborn that creates a box plot. 
# A box plot is a standardized way of displaying the distribution of data based on a five-number summary: 
# minimum, first quartile (Q1), median, third quartile (Q3), and maximum. 
# It's1 useful for identifying outliers and comparing distributions between groups.
```

```python
vis3 = sns.lmplot(data=gdp, x = "InternetUsers", y = "BirthRate", fit_reg = False)
plt.show() 
#  is a function in Seaborn that creates a linear model plot. 
# It's primarily used to visualize the relationship between two numerical variables, often with the option to add a regression line.

# fit_reg controls whether a linear regression line (a line of best fit) is added to the plot.

# fit_reg=False means that no regression line will be drawn. 
# This results in a simple scatter plot showing the relationship between "InternetUsers" and "BirthRate". 
# If this was set to true, a regression line would have been added.
```

```python
vis3 = sns.lmplot(data=gdp, x = "InternetUsers", y = "BirthRate")
plt.show() 
```

```python
vis3 = sns.lmplot(data=gdp, x = "InternetUsers", y = "BirthRate", fit_reg = True)
plt.show()
```

```python
vis5 = sns.lmplot(data = gdp, x = "InternetUsers", y = "BirthRate",fit_reg = False, hue = "IncomeGroup")
plt.show()
```

```python
vis5 = sns.lmplot(data = gdp, x = "InternetUsers", y = "BirthRate", hue = "IncomeGroup")
plt.show()
```

```python
vis5 = sns.lmplot(data = gdp, x = "InternetUsers", y = "BirthRate", hue = "IncomeGroup", height = 10)
plt.show() # size has been replaced with height and aspect in seaborn
```

```python
vis5 = sns.lmplot(data = gdp, x = "InternetUsers", y = "BirthRate", hue = "IncomeGroup", aspect = 2)
plt.show()
# The aspect argument controls the width of the plot relative to its height.
# aspect = 1 means the width will be equal to the height.
# aspect = 2 means the width will be twice the height, and so on.
```
